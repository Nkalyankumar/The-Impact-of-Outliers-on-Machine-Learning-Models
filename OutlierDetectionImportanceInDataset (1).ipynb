# Detection and Elimination of Outliers 

# 1.Detecting Outliers in the DataSet

```python
import numpy
#Removing The Outlier from the Dataset
import pandas

outlier_indices_array = []
dataframe = pandas.read_csv('salary_data.csv')

Q1 = numpy.percentile(dataframe['YearsExperience'],25)
Q2 = numpy.percentile(dataframe['YearsExperience'],50)
Q3 = numpy.percentile(dataframe['YearsExperience'],75)
IQR = Q3 - Q1

lower_range = Q1 - IQR*1.5
higher_range = Q3 + IQR*1.5

outlier_indices_array.append(numpy.where(dataframe['YearsExperience']>=higher_range)[0])
outlier_indices_array.append(numpy.where(dataframe['YearsExperience']<=lower_range)[0])

Q1 = numpy.percentile(dataframe['Salary'],25)
Q2 = numpy.percentile(dataframe['Salary'],50)
Q3 = numpy.percentile(dataframe['Salary'],75)
IQR = Q3 - Q1

lower_range = Q1 - IQR*1.5
higher_range = Q3 + IQR*1.5


outlier_indices_array.append(numpy.where(dataframe['Salary']>=higher_range)[0])
outlier_indices_array.append(numpy.where(dataframe['Salary']<=lower_range)[0]) 


#converting the array of numpy ndarrays to 1d array
outlier_indices_array = numpy.concatenate(outlier_indices_array).ravel()
#Removing the Duplicate row indices 
outlier_indices_array = list(set(outlier_indices_array))
print("The Rows that are Outliers = ",outlier_indices_array)

#Removing the Rows thatcontain outliers
dataframe.drop(index=outlier_indices_array, inplace=True,axis=0)
```
